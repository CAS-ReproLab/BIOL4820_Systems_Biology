# How to Perform Dimensional Analysis

Dimensional analysis is a powerful method to check for consistency in equations, derive scaling laws, make educated guesses, and reduce the number of variables by grouping them into dimensionless combinations. Below is a step-by-step guide on how to perform dimensional analysis:


## 1. Identify the Fundamental Dimensions

1. **Common Fundamental Dimensions**: 
   - $L$ for length in meters (m)
   - $M$ for mass in kilograms (kg)
   - $T$ for time in seconds (s)
   - $V$ for volume in liters (L)
   - $Q$ for electric charge (sometimes used, depending on context) in Amperes (A)
   - $\Theta$ for temperature (sometimes used) in Kelvin (K; on the absolute scale) or Celsius (C) on the relative scale. K = C + 273.15
   - $N$ for amount of substance in moles (mol)

2. **Assign Dimensions to Variables**: 
   For each physical quantity (e.g., velocity, force, energy), express it in terms of its fundamental dimensions.  
   - Example: Velocity $v$ has dimensions of distance per time or $L T^{-1}$.

## 2. Express the Equation or Physical Relation

1. Write down the physical relation or equation you are analyzing.
   - Example: A physical law might state $ F = m \cdot a $, where $F$ is force, $m$ is mass, and $a$ is acceleration.

2. Convert each term into its fundamental dimensions:
   - Force $F$ has dimensions $ M L T^{-2} $.  
   - Mass $m$ has dimensions $M$.  
   - Acceleration $a$ has dimensions $L T^{-2}$.  

3. Check that both sides of the equation have the same overall dimensions.

## 3. Set Up the Dimensional Consistency and Solve

For the equation to be dimensionally consistent, the dimensions on both sides must match exactly:
$$
[ \text{Left Side} ] = [ \text{Right Side} ].
$$

- If an equation is dimensionally inconsistent, it indicates the equation is either incomplete or incorrect.

## 4. Interpret Results

1. **Check the Form of Your Equation**: Once you have dimensionless groups, your physical law can often be written more simply in terms of these groups.
2. **Guide Experiments and Simulations**: Fewer dimensionless parameters means fewer independent variables to explore.
3. **Use Dimensionless Numbers**: Many well-known numbers are dimensionless, allowing comparisons across different systems with varying scales. You've probably seen these before, but didn't realize where they come from. (See step 5 below for more information).


## 5. Extension: Dimensionless Groups

Dimensionless groups help us set up equations that provide biological insight in a manner that is robust to the scale of observations that we typically make in the lab. This concept will likely be a bit intimidating at first. But stick with it. It's a superpower!

1. **Count the Number of Fundamental Dimensions ($k$)**: Determine how many independent fundamental dimensions are present in the problem.
2. **Count the Number of Variables ($n$)**: List all the physical variables involved in the problem.
3. **Form $n - k$ Independent Dimensionless Groups**: Combine variables into products or ratios such that each group is dimensionless.  
   - Example: If you have a characteristic velocity $v$, a characteristic length $L$, and a characteristic time $t$, you can form a dimensionless group such as $ \pi_1 = \frac{v\, t}{L} $, which has no net dimensional unit.

By reducing a problem to a smaller set of dimensionless groups, you can simplify experiments and computations because you only need to vary a smaller set of independent dimensionless parameters. This is a more advanced concept, but is very useful because it connects dimensional analysis to basic mathematical modeling procedures more broadly. If this is something that interests you, I highly recommend Batschelet's [Introduction to Mathematics for Life Scientists](https://link.springer.com/book/10.1007/978-3-642-96080-2). It is an amazing survey of quantitative thinking with a wealth of examples. 


### Key Points to Remember

- Always ensure the overall dimensions on each side of an equation match.
- Carefully track and convert units when performing numerical calculations.
- When in doubt, the units will always guide you...

# Example 1: Number of Water Molecules in an 8 fl. oz Glass of Water

In the previous explanation, we listed steps for performing dimensional analysis. Let’s apply each step to find out how many water molecules are in an 8 fl. oz glass of water.

## 1. Identify the Fundamental Dimensions
- **Length (L)**, **Mass (M)**, **Time (T)**, etc.
  
For our problem, we mainly need to track volume and mass. We’ll be using units such as fluid ounces for volume and grams for mass.

## 2. Express the Physical Relation

**Goal**: Convert 8 fl. oz of water into number of water molecules. There isn't a well known formula for this, it's really just multiplication...

We'll need some additional information though. 

1. the density of water, which relates volume to mass is 1 g/mL (which is equivalent to 1 mL/g).
2. the molar mass of water (in grams/mole). This can be obtained from the periodic table. The molar mass of a molecule is the sum of the atomic masses that make up the molecule. So, in the case of $H_2O$, there are two hydrogens with 1.00 gram/mole and one oxygen with 16 grams/mole for a total of 18 grams/mole.
3. the number of unit substance in a mole. This is given by avogadro's number $N_A= 6.022 * 10^{23}$. The definition of 'unit substance' will depend on the problem. In this context, it is molecules/mole. 

## 3. Set Up Dimensional Consistency

We can chain our conversions in the following way, ensuring consistency:

1. **Convert from fluid ounces to milliliters**  
   - 1 fl. oz $\approx 29.5735 \, \text{mL}$

2. **Convert from milliliters to grams**  
   - Density of water $\approx 1 \, \text{g/mL}$

3. **Convert from grams to moles**  
   - Molar mass of water $\approx 18 \, \text{g/mole}$

4. **Convert from moles to number of molecules**  
   - Avogadro's number $N_A \approx 6.022 \times 10^{23} \, \text{molecules/mole}$

## 4. Perform the Conversions (Step-by-Step)

Let’s follow the sequence of multiplications with their dimensional units:

$$
8 \,\text{fl. oz}
\times \frac{29.5735 \,\text{mL}}{1\,\text{fl. oz}}
\times \frac{1 \,\text{g}}{1\,\text{mL}}
\times \frac{1\,\text{mole}}{18\,\text{g}}
\times \frac{6.022\times10^{23}\,\text{molecules}}{1\,\text{mole}}.
$$

1. **Fluid ounces to milliliters**  
   $$
   8 \,\text{fl. oz} 
   \times 29.5735 \,\frac{\text{mL}}{\text{fl. oz}} 
   \approx 236.588 \,\text{mL}
   $$

2. **Milliliters to grams**  
   $$
   236.588 \,\text{mL} 
   \times 1 \,\frac{\text{g}}{\text{mL}}
   = 236.588 \,\text{g}
   $$

3. **Grams to moles**  
   $$
   236.588 \,\text{g} 
   \times \frac{1\,\text{mole}}{18\,\text{g}}
   \approx 13.1438 \,\text{moles}
   $$

4. **Moles to molecules**  
   $$
   13.1438 \,\text{moles} 
   \times 6.022 \times 10^{23} \,\frac{\text{molecules}}{\text{mole}}
   \approx 7.92\times10^{24}\,\text{molecules}.
   $$

Thus, an 8 fl. oz glass of water contains approximately $7.9 \times 10^{24}$ water molecules.

## Note: How Do We Know How to Set Up the Fractions?

Let's look at this relation: 

1. **Fluid ounces to milliliters**  
   $$
   \frac{8 \,\text{fl. oz}}{1} 
   \times \frac{29.5735 \ \text{mL}}{1 \ \text{fl. oz}} 
   \approx 236.588 \,\text{mL}
   $$

Here, the final units are our desired units when we multiply across the numerator and denominator.

But what if we weren't sure how to set up the conversion of fl. oz to mL? We can set the fraction up in two possible ways... For example, we could also do it this way: 

2. **Fluid ounces to milliliters**  
   $$
   \frac{8 \,\text{fl. oz}}{1} 
   \times \frac{1 \ \text{fl. oz }}{29.5735 \ \text{mL}} 
   \approx \frac{0.271 \ \text{fl .oz}^2}{1 mL}
   $$

Clearly, these are not the units we wanted. So `when in doubt, try it out`. Set up the unit conversions, multiply through and adjust until you get the desired units. 

## 5. Interpret the Results

- **Physical Meaning**: Even a relatively small amount of water (8 fl. oz) contains an extraordinarily large number of molecules — on the order of $10^{24}$!
- **Dimensional Check**: By chaining together these conversions, each intermediate step remains dimensionally consistent, and the final result is a dimensionless count of molecules.

## Here's a python program that performs this calculation

```python
# Python Program
volume_fl_oz = 8
ml_per_fl_oz = 29.5735
density_water_g_per_ml = 1.0
molecular_weight_water_g_per_mole = 18.0
avogadro_number = 6.022e23

volume_ml = volume_fl_oz * ml_per_fl_oz
mass_g = volume_ml * density_water_g_per_ml
moles_water = mass_g / molecular_weight_water_g_per_mole
num_molecules = moles_water * avogadro_number

print(f"Approximate number of water molecules: {num_molecules:.2e}")
```

# Example 2: Estimating the Number of Protein Molecules in a Typical Animal Cell

Below is a rough estimation of how many protein molecules are in a single animal cell. Please note these are highly approximate figures chosen for illustrative purposes only.

## 1. Identify the Fundamental Dimensions

- **Volume**: typically measured in liters (L).
- **Mass**: typically measured in grams (g).
- **Moles**: relates mass to molecular count (mol).

In this example, we'll use the density of water (1 g/mL) and assume the cell is mostly water-like in density. This is a fair approximation as it is probably close to 1.05 g/mL in reality. 

## 2. Express the Physical Relation

1. Start with a **typical cell volume** (e.g., $1 \,\text{pL}$ or $1 \times 10^{-12} \,\text{L}$). I got this figure from a simple google search. The internet said it is around 5 pL, but we'll go with just one here to simplify the calculations. Note: $pL$ is picoliter or 1 trillionth of a liter. 
2. Convert volume to mass (using water density).
3. Estimate the fraction of that mass that is made of protein.
4. Convert protein mass to moles (using the protein’s average molecular weight).
5. Convert moles of protein to number of protein molecules (using Avogadro’s number).

## 3. Set Up the Dimensional Consistency

Our overall dimensional chain:
$$
\underbrace{V_{cell}}_{\text{pL}}
\times
\frac{\text{g}}{\text{mL}}
\times
(\text{fraction of protein})
\times
\frac{\text{mole}}{\text{g}}
\times
\frac{\text{molecules}}{\text{mole}}
=
\underbrace{\text{number of protein molecules}}_{\text{dimensionless count}}
$$

## 4. Perform the Conversions Step-by-Step

1. **Cell Volume**  
   - Assume each cell has a volume of about $1 \,\text{pL} = 1 \times 10^{-12} \,\text{L}$.

2. **Convert Volume to Mass**  
   - $1 \,\text{L}$ of water has a mass of $1000 \,\text{g}$.  
   - Therefore, $1 \,\text{pL} = 1 \times 10^{-12} \,\text{L} \times 1000 \,\frac{\text{g}}{\text{L}} = 1 \times 10^{-9} \,\text{g}$.

3. **Estimate Protein Fraction**  
   - Let’s assume $\approx 15\%$ of the cell’s mass is protein.  
   $$
   1 \times 10^{-9} \,\text{g} \times 0.15 = 1.5 \times 10^{-10} \,\text{g} \,(\text{protein}).
   $$

4. **Convert Protein Mass to Moles**  
   - Assume an **average molecular weight** of proteins is about $50{,}000 \,\text{g/mol}$ (50 kDa).  
   $$
   \frac{1.5 \times 10^{-10} \,\text{g}}{5 \times 10^{4} \,\frac{\text{g}}{\text{mole}}}
   = 3.0 \times 10^{-15} \,\text{moles}.
   $$

**NOTE: the [Dalton](https://en.wikipedia.org/wiki/Dalton_(unit)), abbrieviated above as kilodalton kDa, is another name for the grams/mole mass unit.

5. **Convert Moles to Number of Molecules**  
   - Use **Avogadro’s number** $N_A \approx 6.022 \times 10^{23} \,\text{molecules/mole}$.  
   $$
   3.0 \times 10^{-15} \,\text{moles} 
   \times
   6.022 \times 10^{23} \,\frac{\text{molecules}}{\text{mole}}
   \approx
   1.81 \times 10^{9} \,\text{molecules}.
   $$

Putting it all together:
$$
1 \,\text{pL} \times \frac{1000\,\text{g}}{1\,\text{L}}
\times 0.15
\times
\frac{1\,\text{mole}}{5 \times 10^{4}\,\text{g}}
\times
6.022 \times 10^{23}\,\frac{\text{molecules}}{\text{mole}}
\approx
1.8 \times 10^{9}\,\text{protein molecules}.
$$

## 5. Interpret the Results

- A **single animal cell** of approximately 1 pL volume may contain on the order of **$10^9$ protein molecules** (about 1.8 billion), assuming 15% of its mass is protein with an average molecular weight of 50 kDa.
- Real cells have **varying volumes**, protein concentrations, and a wide range of protein molecular weights. Thus, the exact number can differ significantly between cell types.
- This calculation illustrates how dimensional analysis provides a consistent framework for combining multiple physical quantities and assumptions into a single, approximate estimate.
- I hope you're starting to see that you can get pretty far with a little imagination, educated guessing, and the internet. 

## Here's a Python Program that performs this calculation

```python
# Optional Python Calculation

# Assumptions
volume_cell_pL = 1.0          # cell volume in pL
fraction_protein = 0.15       # fraction of cell mass that is protein
avg_protein_mw_g_per_mol = 5.0e4  # average molecular weight (50 kDa)
avogadro_number = 6.022e23    # molecules per mole

# Convert volume from pL to L
volume_cell_L = volume_cell_pL * 1e-12

# Convert volume (L) of water to grams (assuming density of 1 g/mL)
# 1 L of water => 1000 g of water
cell_mass_g = volume_cell_L * 1000

# Fraction of mass that is protein
protein_mass_g = cell_mass_g * fraction_protein

# Convert protein mass to moles
moles_protein = protein_mass_g / avg_protein_mw_g_per_mol

# Convert moles to number of molecules
num_protein_molecules = moles_protein * avogadro_number

print(f"Approximate number of protein molecules in one cell: {num_protein_molecules:.2e}")
```

# Example 3: Time to Fall from a Given Height

Let's look at a simpler illustration of how to use dimensional analysis to find the time it takes an object to fall a certain distance under gravity (with no air resistance). 

**Step 1: Identify the Variables and Their Dimensions**

We want the **time to fall** $(T)$ from a height $(H)$ under gravitational acceleration $(g)$.

| Variable | Description                         | Dimensions       |
|----------|-------------------------------------|------------------|
| $T$      | Time to fall                        | $T$              |
| $H$      | Height                              | $L$              |
| $g$      | Gravitational acceleration          | $L\,T^{-2}$      |

**Step 2: Write the General Relation**

We hypothesize that freefall time (T) is a `function` of height (H) and gravity (g):

$$
T = f(H, g).
$$

**Step 3: Count the Number of Variables and Fundamental Dimensions**

- **Number of variables ($n$)**: 3 ($T$, $H$, $g$)
- **Number of fundamental dimensions ($k$)**: 2 ($L$, $T$)

By the **Buckingham $\pi$-Theorem**:

$$
\text{Number of dimensionless groups} = n - k = 3 - 2 = 1.
$$

Hence, we can form **one** dimensionless group.

**Step 4: Form the Dimensionless Group**

Let’s propose a group $\pi_1$. We write:

$$
\pi_1 = T^a \, H^b \, g^c.
$$

Next, insert the dimensions:

- $T$ has dimension $T$.
- $H$ has dimension $L$.
- $g$ has dimension $L\,T^{-2}$.

In terms of fundamental dimensions:

$$
[T]^a \; [L]^b \; [L\,T^{-2}]^c 
= T^a \; L^{b + c} \; T^{-2c}.
$$

To be dimensionless, the exponents of $L$ and $T$ must both be zero:

1. **Exponent on $L$**: $b + c = 0$.  
2. **Exponent on $T$**: $a - 2c = 0$.

Let’s solve:

- From (2): $a = 2c$.  
- From (1): $b = -c$.

Choose $c = 1$ for convenience. Then:

$$
a = 2, \quad b = -1, \quad c = 1.
$$

Therefore:

$$
\pi_1 = T^2 \, H^{-1} \, g^1 
      = \frac{g \, T^2}{H}.
$$

**Step 5: Rewrite the Original Equation**

$$
\pi_1 = \frac{g \, T^2}{H} 
\quad \Longrightarrow \quad 
\frac{g \, T^2}{H} = \text{constant} \ (\kappa).
$$

Hence,

$$
g \, T^2 = \kappa \, H 
\quad \Longrightarrow \quad
T^2 = \kappa \, \frac{H}{g}.
$$

Taking the square root:

$$
T = \sqrt{\kappa \,\frac{H}{g}}.
$$

In a classic free-fall derivation, ignoring air resistance, physics shows:

$$
T = \sqrt{\frac{2 H}{g}}
\quad \Longrightarrow \quad
\kappa = 2.
$$

How we arrive at this conclusion is detailed in the next section.

**Step 6: Interpret the Result**

1. **Dimensionless Group**  
   $\displaystyle \frac{g T^2}{H}$ is dimensionless. If two falling objects (in different gravitational fields or from different heights) share the same value of $\frac{g T^2}{H}$, they exhibit the same **dimensionless** fall behavior.

2. **Scaling Law**  
   $T$ scales like $\sqrt{\tfrac{H}{g}}$. Increasing the drop height $H$ makes it take longer to fall, and increasing $g$ makes it faster.

3. **Physical Insight**  
   We recover the familiar result $T = \sqrt{\tfrac{2H}{g}}$. The dimensional-analysis approach shows how the **scaling** emerges without needing the exact numerical constant (2) until the final step.

## Here's a Python Program to Simulate Experimental Determineation of k

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_fall_times(num_points=50, kappa_true=2.0, g=9.81, 
                        min_height=0.1, max_height=10.0, noise_level=0.02):
    """
    Simulate fall times T for various heights H under gravity g.
    The true model is T = sqrt(kappa_true * H / g) + noise.
    
    Parameters
    ----------
    num_points : int
        Number of data points to simulate.
    kappa_true : float
        The actual proportionality constant (kappa) used in the simulation. This is well-known in physics.
    g : float
        Gravitational acceleration (m/s^2).
    min_height, max_height : float
        Range of heights (in meters) from which to sample.
    noise_level : float
        Standard deviation of Gaussian noise to add to the simulated times.
        
    Returns
    -------
    H : ndarray
        Array of heights (m).
    T : ndarray
        Array of simulated fall times (s).
    """
    # Random heights between min_height and max_height
    H = np.random.uniform(low=min_height, high=max_height, size=num_points)
    
    # Theoretical noise-free time: T = sqrt(kappa * H / g)
    T_ideal = np.sqrt(kappa_true * H / g)
    
    # Add Gaussian noise to simulate measurement error
    noise = np.random.normal(loc=0.0, scale=noise_level, size=num_points)
    T_measured = T_ideal + noise
    
    return H, T_measured

def estimate_kappa(H, T, g=9.81):
    """
    Estimate the proportionality constant kappa from data using
    the relationship: T^2 = kappa * (H / g).
    
    Rearrange to: kappa = (T^2 * g) / H.
    We then take an average over all data points (or could do a regression).
    
    Parameters
    ----------
    H : ndarray
        Array of heights (m).
    T : ndarray
        Array of measured fall times (s).
    g : float
        Gravitational acceleration (m/s^2).
        
    Returns
    -------
    kappa_est : float
        Empirical estimate for the proportionality constant.
    """
    # Avoid divide-by-zero if any H are zero
    valid = H > 0
    # kappa_i = (T_i^2 * g) / H_i
    kappa_values = (T[valid]**2 * g) / H[valid]
    
    # Simple approach: take the mean
    kappa_est = np.mean(kappa_values)
    return kappa_est


# 1. Simulate data
np.random.seed(42)  # for reproducible results
num_points = 50
kappa_true = 2.0
g = 9.81
    
H, T_measured = simulate_fall_times(
    num_points=num_points,
    kappa_true=kappa_true,
    g=g,
    min_height=0.1,
    max_height=10.0,
    noise_level=0.07
)
    
# 2. Estimate kappa from the simulated data
kappa_est = estimate_kappa(H, T_measured, g=g)
    
# 3. Print results
print(f"True kappa = {kappa_true:.3f}")
print(f"Estimated kappa = {kappa_est:.3f}")
    
# 4.Plot the data
T_theory = np.sqrt(kappa_est * H / g)
plt.figure(figsize=(8, 5))
plt.scatter(H, T_measured, color='blue', label='Simulated data')
plt.scatter(H, T_theory, color='red', label='Fit using estimated kappa')
plt.xlabel('Height (m)')
plt.ylabel('Fall Time (s)')
plt.title('Estimated vs. Simulated Fall Times')
plt.legend()
plt.grid(True)
plt.show()
```

## Determining the True Value of k by Integrating Velocity (i.e., Using Calculus)

In the plot above, the red dots represent the theoretical relationship with freefall time as a function of height and gravity. I mentioned in the previous section that the theoretical value of $k$ is 2. We can get there through dimensional analysis and experiments, or as it turns out, we can also get there through clever analysis using calculus. That's pretty awesome, but the dirty secret of calculus is that it's rare and often quite difficult to find these kinds of analytical solutions using identities to solve the integrals. The former method, using dimensional analysis and experiments, works more often. 

Now, to see why $kappa$ (the proportionality constant) equals 2 in the ideal free-fall problem, we start with Newton’s second law under constant gravitational acceleration $g$. For an object dropped from rest at height $H$, the acceleration $a$ is $g$ in the downward direction. This means the velocity $v(t)$ increases linearly with time:

$$
\frac{dv}{dt} = g 
\quad\Longrightarrow\quad
v(t) = g\,t \quad (\text{assuming }v(0)=0).
$$

The position $x(t)$ (measured from the starting point downward) is then given by integrating $v(t)$ with respect to time:

$$
x(t) 
= \int_{0}^{t} v(\tau)\, d\tau 
= \int_{0}^{t} g\,\tau \, d\tau 
= g\,\frac{t^2}{2}.
$$

At the moment the object reaches the ground ($x = H$) after time $T$, we have:

$$
H = g\,\frac{T^2}{2}
\quad\Longrightarrow\quad
T^2 = \frac{2\,H}{g}.
$$

Comparing this with the dimensionless form $T^2 = kappa\,\frac{H}{g}$ shows that

$$
kappa = 2.
$$

Thus, the constant 2 arises directly from integrating the linearly increasing velocity over time when acceleration is constant $g$ and the initial velocity is zero.

# Example 4: Scaling of Oxygen Diffusion in Cells

This concept of dimensionless groups is probably a bit more challenging than you are used to. That's totally OK. Just keep at it. I promise it's worth it... It's a superpower. 

**Problem Statement**

The rate at which oxygen diffuses into a cell depends on:
1. The **diffusion coefficient** of oxygen in the cytoplasm ($D$) with dimensions $L^2 T^{-1}$,
2. The **cell's radius** ($R$) with dimensions $L$,
3. The **oxygen concentration gradient** across the cell ($\Delta C$) with dimensions $M L^{-3}$.

As we compare cells of different sizes, the diffusion of oxygen will be slower in larger cells. Let's imagine the cell is a sphere. Its volume is: 

$$
V= \frac{4}{3} \pi \cdot R^3 
$$

The radius (R) has fundamental units of length ($L$). It is cubed (raised to the power of 3, because there is an x, y, and z direction in 3D and a sphere is a 3D shape. So, in a sphere, there are three different directions the length (R) will go, but they are all the same, so the fundamental unit is $L$.

Intuitively, a unit increase in radius, would increase the size of the sphere in all 3 directions. Thus, the volume is proportional to the cubed radius. In other words, a small change in radius translates to a large change in volume. We would predict that a small change in radius would also have a large accompanying change in the rate of diffusion of gases.

What we want, is an equation that can help us relate this effect to the physical process of gas diffusion into the cell. 

Now, let's explore the utility of dimensionless groups by analyzing how the rate of oxygen diffusion into a cell scales with its size.

**Step 1: Identify the Variables and Their Dimensions**- You can this information from the internet.

| Variable       | Description                               | Dimension          | Unit       |
|----------------|-------------------------------------------|--------------------|------------|
| $J$           | Oxygen diffusion rate (flux)              | $M L^{-2} T^{-1}$ | $g/m^{2}s$|
| $D$           | Diffusion coefficient                     | $L^2 T^{-1}$      | $m^2/s$   |
| $R$           | Radius of the cell                        | $L$               | $m$       |
| $\Delta C$    | Oxygen concentration gradient across cell | $M L^{-3}$        | $g/m^3$   |

---

**Step 2: Write the General Relation**

Assume the diffusion rate $J$ depends on $D$, $R$, and $\Delta C$. We would say that $J$ is a `function` of $D$, $R$, and $\Delta{C}$:
$$
J = f(D, R, \Delta C)
$$

---

**Step 3: Count the Number of Variables and Dimensions**

- Number of variables ($n$): 4 ($J$, $D$, $R$, and $\Delta C$),
- Number of fundamental dimensions ($k$): 3 ($M$, $L$, and $T$). We can get that from the table above. 

We will now use a useful little concept called: [Buckingham $\pi$-Theorem](https://en.wikipedia.org/wiki/Buckingham_%CF%80_theorem):


This theorem is a method for determining the number of dimensionless groups we can make based on the number of variables, and the number of fundamental dimensions. This calculation is pretty straightforward using the info above:

$$
\text{Number of dimensionless groups} = n - k = 4 - 3 = 1
$$

So, we can form one dimensionless group.

---

**Step 4: What the Heck is a Dimensionless Group?**

To make a dimensionless group, we must essentially rearange the variables in the equation based on the `redundancy` in their fundamental dimensions. This will give us one number (in this specific example), that has no fundamental units (i.e., is dimensionless) and represents proportionality of the equation regardless of the specific scale that we are looking at. 

For example, we may observe gas diffusion at short length scales by making measurements in cells that have very small differences in volume, or we might observe cells that have very large volume differences such as bacteria Vs. animal cells. What we want to know is how this behavior is related across scales. The dimensionaless group allows us to define a constant of proportionality that achieves that by making our equation depend less on the scale. We will denote our dimensionless group here as $\pi_1$. 

Let:
$$
\pi_1 = J^a D^b R^c \Delta C^d
$$

Now, the next step is to substitute the dimensions of each variable (See math appendix A for a refresher on exponent rules):
$$
[M^{1} L^{-2} T^{-1}]^a [L^2 T^{-1}]^b [L^{1}]^c [M^{1} L^{-3}]^d = M^{1a+1d} L^{-2a+2b+c-3d} T^{-1a-1b}
$$

For $\pi_1$ to be dimensionless, the exponents of $M$, $L$, and $T$ must be zero:
1. For $M$: $a + d = 0$,
2. For $L$: $-2a + 2b + c - 3d = 0$,
3. For $T$: $-a - b = 0$.

Solve these equations:
- From (3): $b = -a$,
- From (1): $d = -a$,
- Substitute $b = -a$ and $d = -a$ into (2): 
  $$
  -2a + 2(-a) + c - 3(-a) = 0 \implies -2a - 2a + c + 3a = 0 \implies c = a.
  $$

Let $a = 1$ (arbitrary scaling choice), so:
$$
b = -1, \quad c = 1, \quad d = -1.
$$

Remember this?:

$$
\pi_1 = J^a D^b R^c \Delta C^d
$$

By substituting $a$, $b$, $c$, and $d$ we have: 

$$
\pi_1 = J^1 D^{-1} R^1 \Delta C^{-1}
$$

Another way to write that is:
$$
\pi_1 = \frac{J R}{D \Delta C}
$$

By convention, this is rewritten as:
$$
\text{Dimensionless group: } \frac{J R}{D \Delta C}
$$

---

**Step 5: Rewrite the Equation**

The final version of the equation we want is in terms of the diffusion (flux; J). So let's rearrange accordingly.

Rearranging $\pi_1 = \frac{J R}{D \Delta C}$, we get:
$$
J = \pi_1 \cdot \frac{D \Delta C}{R}
$$

Now, we can rename $\pi_1$ with $k$, a much more commonly used term for a constant of poportionality. 

$$
J = k \cdot \frac{D \Delta C}{R}
$$

where $k$ is a dimensionless constant.

---

**Step 6: Interpret the Result**

1. **Scaling Law**:
   - The oxygen flux $J$ scales inversely with cell radius $R$, meaning larger cells have lower diffusion rates per unit volume if all else remains constant.

2. **Utility**:
   - This dimensionless relationship helps predict diffusion limits in larger cells or organelles.
   - The $\pi_1$ group can be used to compare oxygen flux across cells of different sizes, diffusion coefficients, or concentration gradients.
   - We would need to determine this value using experiments and plotting the outcomes.

3. **Biological Insight**:
   - Cells may adapt to diffusion constraints by increasing surface area-to-volume ratios (e.g., having invaginated surfaces) or developing mechanisms like active transport to compensate for the limitations of diffusion.

# Some Additional Dimensional Analysis Examples (With Solutions)

Below are five scenarios from cell biology and biochemistry that highlight checking the validity of equations using dimensional analysis. They may look a bit frightening at first, but all the are asking is whether the fundamental units in the equations are correctly associated with the solution. 

You don't need to solve these. Solutions are provided. The goal is to give you many examples to see the application of dimensional analysis in different contexts. 

## 1. Estimating the Diffusion Time of a Small Molecule Across a Cell

- **Scenario**: A small signaling molecule (e.g., Ca²⁺) diffuses from one side of the cell to the other.  
- **Dimensions / Assumptions**:
  - Cell diameter (characteristic length): $D$ (meters, $L$)
  - Diffusion coefficient: $D_{\text{diff}}$ ($\mathrm{m^2/s}$, $L^2 T^{-1}$)

- **Characteristic Equation**:  
  A common approximation for the characteristic **diffusion time** $t$ is  
  $$t \sim \frac{D^2}{D_{\text{diff}}}.$$

- **Goal**:  
  *Use dimensional analysis to confirm that $t$ has units of time and to understand how $D$ and $D_{\text{diff}}$ scale the timescale for diffusion.*

**Solution**
1. **Dimensions of the numerator**:  
   - $D^2 \rightarrow (L)^2 = L^2$.
2. **Dimensions of the denominator**:  
   - $D_{\text{diff}} \rightarrow L^2 T^{-1}$.
3. **Ratio**:  
   $$\frac{D^2}{D_{\text{diff}}}
     = \frac{L^2}{\,L^2\,T^{-1}}
     = T.$$
   Hence, $\frac{D^2}{D_{\text{diff}}}$ indeed has units of time, confirming $t$ is a timescale.

4. **Physical Interpretation**:  
   - Larger cell size ($D$) means $t$ grows as $D^2$.  
   - A higher diffusion coefficient $D_{\text{diff}}$ (faster diffusion) **decreases** $t$.


## 2. Calculating the Copy Number of a Newly Expressed Protein

- **Scenario**: A cell synthesizes a protein at a certain rate; you want to estimate how many protein molecules accumulate in a given time.  
- **Dimensions / Assumptions**:
  - Cell volume: $V_{\text{cell}}$ (e.g., liters or $\mathrm{m^3}$) — though not always needed for a simple count.
  - Rate of protein synthesis: $r_s$ (moles per second, $\mathrm{mol\,s^{-1}}$) or (moles per liter per second, $\mathrm{mol\,L^{-1}\,s^{-1}}$).
  - Duration of synthesis: $\Delta t$ (seconds, $T$)
  - Average molecular weight of the protein: $M_p$ (g/mol, $M$ in dimensional notation)
  - Avogadro’s number: $N_A \approx 6.022 \times 10^{23}\,(\text{molecules/mol})$ (dimensionless multiplier from “moles” to “count of molecules”)

- **Characteristic Equations**:  
  1. **Total moles of protein** produced after time $\Delta t$:  
     $$n_p = r_s \times \Delta t 
       \quad (\text{if } r_s \text{ is in moles/s}).$$
  2. **Number of molecules**:  
     $$N_{\text{molecules}} = n_p \times N_A.$$

- **Goal**:  
  *Use dimensional analysis to confirm these equations and the final count of protein molecules.*

**Solution**
1. **Total moles**:
   - $r_s$ has dimension $\mathrm{mol\,s^{-1}}$.
   - $\Delta t$ has dimension $T$.
   - Multiplying:
     $$\mathrm{(mol/s)} \times \mathrm{(s)} = \mathrm{mol}.$$
   - So $n_p$ is in moles, as expected.

2. **Number of molecules**:
   - $n_p$ in moles $\rightarrow$ multiply by Avogadro’s number (molecules/mole) $\rightarrow$ dimensionless “count of molecules.”
   - So $N_{\text{molecules}}$ is a pure number (the actual count).

This confirms that $\Delta t$ and $r_s$ produce the correct dimension for moles and that multiplying by $N_A$ yields the final number of protein molecules.


## 3. Determining the Rate of Metabolite Transport Across a Cell Membrane

- **Scenario**: A particular metabolite (e.g., glucose) crosses the cell membrane (area $A_{\text{mem}}$) by passive diffusion with permeability $P$ under a concentration difference $\Delta C$.  
- **Dimensions / Assumptions**:
  - Membrane surface area: $A_{\text{mem}}$ ($L^2$)
  - Permeability: $P$ ($L\,T^{-1}$) or more precisely $\mathrm{m/s}$
  - Concentration difference: $\Delta C$ ($\mathrm{mol\,m^{-3}}$) or ($\mathrm{g\,m^{-3}}$), depending on usage

- **Characteristic Equation** (for passive permeation flux $J$ in $\mathrm{mol/s}$ or $\mathrm{g/s}$):  
  $$J = P \, A_{\text{mem}} \,\Delta C.$$

- **Goal**:  
  *Use dimensional analysis to ensure $J$ ends up in the correct units of mass or moles per second.*

**Solution**
1. **Permeability** $P$: $\mathrm{m/s}$ ($L\,T^{-1}$).
2. **Membrane area** $A_{\text{mem}}$: $\mathrm{m^2}$ ($L^2$).
3. **Concentration difference** $\Delta C$: $\mathrm{mol\,m^{-3}}$ ($M\,L^{-3}$) or $\mathrm{g\,m^{-3}}$ depending on mass vs. moles.
4. **Multiply them**:
   $$P \times A_{\text{mem}} \times \Delta C
     = (L\,T^{-1})
       \times (L^2)
       \times (M\,L^{-3})
     = M \, T^{-1}.$$
   - If $M$ represents moles, then $M\,T^{-1} = \mathrm{mol\,s^{-1}}$.
   - If $M$ represents mass, then $M\,T^{-1} = \mathrm{g\,s^{-1}}$. (where is grams)

$J$ is in the correct units of mass flow rate or molar flow rate (mass/time or moles/time).


## 4. Estimating the Time for Enzymatic Conversion of a Substrate in a Cell Extract

- **Scenario**: An enzyme at concentration $[E]$ catalyzes a reaction converting substrate $[S]$ to product.  
- **Dimensions / Assumptions**:
  - Initial substrate concentration: $[S_0]$ (mol/L)
  - Enzyme concentration: $[E]$ (mol/L)
  - Catalytic turnover rate: $k_{\text{cat}}$ ($\mathrm{s^{-1}}$)
  - Assume all substrate eventually gets converted and the enzyme is not consumed (simple model).

- **Characteristic Equation** (simplified model):  
  A rough estimate for the **completion time** $t_{\text{rxn}}$ might be given by  
  $$t_{\text{rxn}} \sim \frac{[S_0]}{k_{\text{cat}} \,[E]}.$$

- **Goal**:  
  *Perform a dimensional analysis check to verify that the ratio $\frac{[S_0]}{k_{\text{cat}} [E]}$ yields a time.*

**Solution**
1. **Substrate concentration** $[S_0]$: $\mathrm{mol\,L^{-1}}$.
2. **Enzyme concentration** $[E]$: $\mathrm{mol\,L^{-1}}$.
3. **Catalytic rate** $k_{\text{cat}}$: $\mathrm{s^{-1}}$.
4. **Denominator**:
   $$k_{\text{cat}} \,[E]
     = \mathrm{s^{-1}}
       \times \mathrm{mol\,L^{-1}}
     = \mathrm{mol\,L^{-1}\,s^{-1}}.$$
5. **Overall ratio**:
   $$\frac{[S_0]}{k_{\text{cat}} [E]}
     = \frac{\mathrm{mol\,L^{-1}}}
            {\mathrm{mol\,L^{-1}\,s^{-1}}}
     = \mathrm{s}.$$

Therefore, the units are time.


## 5. Determining the Sedimentation Velocity of Organelles in a Centrifuge

- **Scenario**: Cellular organelles (e.g., mitochondria) are separated by spinning in a centrifuge. You want to estimate the velocity at which they move radially outward.  
- **Dimensions / Assumptions**:
  - Angular velocity of the rotor: $\omega$ ($\mathrm{rad\,s^{-1}} \approx \mathrm{s^{-1}}$) (Note: rad is a radian unit. A complete revolution is equal to $2\cdot \pi$ radians. 
  - Distance from the rotor axis: $r$ ($L$)
  - Mass of the organelle: $m$ ($M$, or $\mathrm{kg}$)
  - Drag coefficient: $\gamma$ ($\mathrm{kg\,s^{-1}}$) — how much friction do the migrating organelles experience in the mixture being centrifuged.

- **Characteristic Equation** (radial equilibrium):
  $$m\,\omega^2\,r = \gamma\,v \quad\Longrightarrow\quad
    v = \frac{m\,\omega^2\,r}{\gamma}.$$

- **Goal**:  
  *Use dimensional analysis to confirm that the above relation produces a velocity ($\mathrm{m/s}$).*

**Solution**
1. **Left side** (centrifugal force): $m\,\omega^2\,r$
   - $m$ in $\mathrm{kg}$ ($M$).
   - $\omega^2 \sim \mathrm{s^{-2}}$.
   - $r$ in $\mathrm{m}$ ($L$).
   - Combined dimension: $\mathrm{kg \cdot m \cdot s^{-2}} = \mathrm{N}$, which is a force.

2. **Right side** (drag force): $\gamma\,v$
   - $\gamma$ in $\mathrm{kg\,s^{-1}}$.
   - $v$ in $\mathrm{m\,s^{-1}}$ if dimensionless analysis is correct.
   - Product dimension: $(\mathrm{kg\,s^{-1}}) \times (\mathrm{m\,s^{-1}}) = \mathrm{kg\,m\,s^{-2}} = \mathrm{N}.$
   - This confirms the force balance is dimensionally consistent.

3. **Solving for $v$**:
   $$v 
     = \frac{m\,\omega^2\,r}{\gamma}
     \quad\Longrightarrow\quad
     \frac{\mathrm{kg \cdot m \cdot s^{-2}}}
          {\mathrm{kg \cdot s^{-1}}}
     = \frac{\mathrm{m}}{\mathrm{s}}.$$

So $v$ is indeed a velocity in $\mathrm{m/s}$.

# Math Appendix A: Exponents, and Scientific Notation

**Exponents**

1. **Definition**:
   - An exponent represents repeated multiplication:
     $$
     a^n = a \times a \times \cdots \times a \, (n \, \text{times})
     $$

2. **Basic Properties of Exponents**:
   - **Product Rule**:
     $$
     a^m \cdot a^n = a^{m+n}
     $$
   - **Quotient Rule**:
     $$
     \frac{a^m}{a^n} = a^{m-n}, \quad \text{for } a \neq 0
     $$
   - **Power Rule**:
     $$
     (a^m)^n = a^{m \cdot n}
     $$
   - **Zero Exponent**:
     $$
     a^0 = 1, \quad \text{for } a \neq 0
     $$
   - **Negative Exponent**:
     $$
     a^{-n} = \frac{1}{a^n}, \quad \text{for } a \neq 0
     $$

3. **Working with Fractions**:
   - Apply exponents to both numerator and denominator:
     $$
     \left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}, \quad \text{for } b \neq 0
     $$

---

**Scientific Notation**

1. **Definition**:
   - A way to express very large or very small numbers as a product of a number and a power of 10:
     $$
     N \times 10^m, \quad 1 \leq N < 10, \quad m \in \mathbb{Z}
     $$

2. **Examples**:
   - $ 5000 = 5 \times 10^3 $
   - $ 0.00042 = 4.2 \times 10^{-4} $

3. **Operations in Scientific Notation**:
   - **Multiplication**:
     Multiply the coefficients and add the exponents:
     $$
     (a \times 10^m) \cdot (b \times 10^n) = (a \cdot b) \times 10^{m+n}
     $$
     Example:
     $$
     (2 \times 10^3) \cdot (3 \times 10^4) = 6 \times 10^{3+4} = 6 \times 10^7
     $$
   - **Division**:
     Divide the coefficients and subtract the exponents:
     $$
     \frac{a \times 10^m}{b \times 10^n} = \frac{a}{b} \times 10^{m-n}
     $$
     Example:
     $$
     \frac{6 \times 10^5}{3 \times 10^2} = 2 \times 10^{5-2} = 2 \times 10^3
     $$
   - **Addition/Subtraction**:
     Ensure the exponents are the same before adding or subtracting the coefficients:
     $$
     (a \times 10^m) + (b \times 10^m) = (a + b) \times 10^m
     $$
     Example:
     $$
     (2 \times 10^3) + (3 \times 10^3) = (2 + 3) \times 10^3 = 5 \times 10^3
     $$
     If the exponents differ, adjust one number:
     $$
     (5 \times 10^4) + (3 \times 10^3) = (5 \times 10^4) + (0.3 \times 10^4) = 5.3 \times 10^4
     $$

4. **Convert to and from Scientific Notation**:
   - Move the decimal point to create a number between 1 and 10, adjusting the exponent accordingly:
     - $ 12345 = 1.2345 \times 10^4 $
     - $ 0.00076 = 7.6 \times 10^{-4} $

# Math Appendix B: Fractions and Logarithms

**Fractions**

1. **Basic Properties**:
   - A fraction is expressed as $ \frac{a}{b} $, where:
     - $ a $ (numerator): The top number.
     - $ b $ (denominator): The bottom number (non-zero).

2. **Operations with Fractions**:
   - **Addition/Subtraction**: Requires a common denominator.
     $$
     \frac{a}{b} + \frac{c}{d} = \frac{ad + bc}{bd}, \quad \frac{a}{b} - \frac{c}{d} = \frac{ad - bc}{bd}
     $$
   - **Multiplication**:
     $$
     \frac{a}{b} \times \frac{c}{d} = \frac{ac}{bd}
     $$
   - **Division**: Multiply by the reciprocal.
     $$
     \frac{a}{b} \div \frac{c}{d} = \frac{a}{b} \times \frac{d}{c} = \frac{ad}{bc}
     $$

3. **Simplification**:
   - Factorize the numerator and denominator and cancel common factors:
     $$
     \frac{12}{16} = \frac{3 \cdot 4}{4 \cdot 4} = \frac{3}{4}
     $$

4. **Reciprocals**:
   - The reciprocal of $ \frac{a}{b} $ is $ \frac{b}{a} $.

---

**Logarithms**

1. **Definition**:
   - A logarithm answers the question: "To what power must the base $ b $ be raised to yield $ x $?". A logarithm is a function that takes a number and returns a new number representing the input as a magnitude of the base.
     $$
     \log_b(x) = y \quad \text{means} \quad b^y = x
     $$

2. **Common Logarithms**:
   - **Base 10 (common logarithm)**: $ \log(x) = \log_{10}(x) $
   - **Base $ e $ (natural logarithm)**: $ \ln(x) = \log_e(x) $

3. **Basic Properties**:
   - **Product Rule**:
     $$
     \log_b(xy) = \log_b(x) + \log_b(y)
     $$
   - **Quotient Rule**:
     $$
     \log_b\left(\frac{x}{y}\right) = \log_b(x) - \log_b(y)
     $$
   - **Power Rule**:
     $$
     \log_b(x^k) = k \cdot \log_b(x)
     $$

4. **Change of Base Formula**:
   - To convert between bases:
     $$
     \log_b(x) = \frac{\log_k(x)}{\log_k(b)}
     $$
   - Example for base 10:
     $$
     \log_2(8) = \frac{\log_{10}(8)}{\log_{10}(2)}
     $$

5. **Special Cases**:
   - $ \log_b(1) = 0 $ (since $ b^0 = 1 $).
   - $ \log_b(b) = 1 $ (since $ b^1 = b $).
